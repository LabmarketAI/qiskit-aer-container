# jupyter_pubsub — Demo Notebook

This notebook demonstrates the Phase 1 WebSocket pub/sub extension.

## How to subscribe from a browser

Open the browser DevTools console while JupyterLab is running and paste:

```js
// Subscribe to the network-graph cell
const ws = new WebSocket('ws://localhost:8888/pubsub/ws/nx-graph');
ws.onmessage = e => console.log('nx-graph:', JSON.parse(e.data));

// Subscribe to the dataframe cell
const ws2 = new WebSocket('ws://localhost:8888/pubsub/ws/df-summary');
ws2.onmessage = e => console.log('df-summary:', JSON.parse(e.data));
```

Then run the cells below — output appears in the console in real time.

## Discovery endpoint

```
GET http://localhost:8888/pubsub/cells
```
Returns active cell names and subscriber counts.

````python
# pubsub: nx-graph
import networkx as nx
import json

G = nx.karate_club_graph()
data = nx.node_link_data(G, edges="links")
# Print as JSON so it's captured by the 'stream' message type
print(json.dumps(data))
````

````python
# pubsub: df-summary
import pandas as pd

df = pd.DataFrame({
    "dept": ["Engineering", "Marketing", "HR", "Finance"],
    "spend": [120_000, 80_000, 40_000, 95_000],
    "headcount": [25, 12, 8, 15],
})
df
````

````python
# Check active subscriptions
import urllib.request, json
with urllib.request.urlopen('http://localhost:8888/pubsub/cells') as r:
    print(json.dumps(json.loads(r.read()), indent=2))
````
